```python
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
data, out = load_iris(return_X_y=True)
```

```python
out
```

```python
dt_model = DecisionTreeClassifier()
 
```

```python
param_grid = {
    'max_depth':[3,7],
    'min_samples_split':[4,9,10]
}

grid = GridSearchCV(dt_model,param_grid,cv=5,scoring='accuracy')
grid.fit(data,out)

print("Best Parameters(Gridsearch): ",grid.best_params_)
print("Best Score: ",grid.best_score_)
```

```python
param_dist = {
    'max_depth':[1,3,5,11,9],
    'min_samples_split':[2,4,8,3]
}

random_search = RandomizedSearchCV(dt_model,param_dist,n_iter=5,cv=5,scoring='accuracy',random_state=42)

random_search.fit(data,out)
print("Best Parameters (RandomizedSearch):", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```
