```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv("insurance_data.csv")

df.head()
```

```python
df.info()
```

```python
print("Check missing\n",df.isnull().sum())
print("Check Stats\n",df.describe())
```

```python
plt.scatter(df['age'],df['bought_insurance'],color = "red")
plt.title("Age vs insurance bought")
plt.xlabel("Age")
plt.ylabel("Insurance bought")
plt.grid(True)
plt.show()
```

```python
x = df[['age']]
y = df['bought_insurance']
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred = model.predict(X_test)
print("Predicted:", y_pred.tolist())
print("Actual:", y_test.tolist())
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:", accuracy)

cm = confusion_matrix(y_test,y_pred)
print("Confusion Matrix:\n", cm)
```
