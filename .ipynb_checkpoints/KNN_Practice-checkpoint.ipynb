```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
x = np.array([[2], [4], [6], [1], [7]])   
y = np.array(['Fail', 'Pass', 'Pass', 'Fail', 'Pass'])
```

```python
print(x)
print(y)
```

```python
model = KNeighborsClassifier(n_neighbors = 3)
```

```python
model.fit(x,y)
```

```python
new_data = [[3]]
prediction = model.predict(new_data)
print(prediction[0])
```

```python
plt.figure(figsize = (8,5))
for i in range(len(x)):
    if y[i] == 'Pass':
        plt.scatter(x[i],0,color = 'green', label = 'Pass' if i == 1 else "")
    else:
        plt.scatter(x[i],0,color = 'red', label = 'Fail' if i == 3 else "")

plt.scatter(new_data,0,color = 'blue',marker = 'x',s=100,label = 'NewData')

plt.xlabel('Hours Studied')
plt.yticks([])
plt.title('KNN - Pass or Fail Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
