```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("housing.csv")
print("Info\n")
df.info()
print("\nHead\n")
df.head()
```

```python
print("\nTail\n")
df.tail()
```

```python
print("\nDescribe\n")
df.describe()
```

```python
df.isnull().sum()
```

```python
df['sqft_living'] = df['sqft_living'].fillna(df['sqft_living'].median()) # taking median and filling the empty places 
df['sqft_lot'] = df['sqft_lot'].fillna(df['sqft_lot'].median()) # taking median and fillling it in the empty places
df['yr_built'] = df['yr_built'].fillna(df['yr_built'].mode()[0]) # filling the empty places eith the  most occuring year
df['city'] = df['city'].fillna(df['city'].mode()[0]) # usinf the most common city to fill up the empty nul spaces 
```

```python
df.isnull().sum()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y %H:%M')

df['year_sold'] = df['date'].dt.year
df['month_sold'] = df['date'].dt.month

df['house_age'] = df['year_sold'] - df['yr_built']

df['price_log'] = np.log1p(df['price'])

df['is_renovated'] = (df['yr_renovated'] != 0).astype(int)

df = df.drop(['date', 'yr_built', 'yr_renovated'], axis=1)
```

```python
df.head()
```

```python
sns.boxplot(df['price'])
```

```python
df = df[df['price'] < df['price'].quantile(0.99)]
```

```python
sns.boxplot(df['price'])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.scatterplot(x='sqft_living', y='price', data=df,hue = "bedrooms",palette='hsv', s=100, alpha=0.7)
plt.title('Relationship between House Size (sqft_living) and Price')
plt.show()

plt.subplot(1,2,2)
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Impact of Number of Bedrooms on House Price')
plt.show()

print("Heat Map of the correlation")
plt.figure(figsize=(10,8))
corr_matrix = df.select_dtypes(include=['int64', 'float64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',fmt = ".2f")
plt.title('Correlation Heatmap of Housing Features')
plt.show()
```
