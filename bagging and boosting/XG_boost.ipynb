```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
dir(housing)
```

```python
import pandas as pd
housing_df = pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
housing_df['MedHouseVal'] = housing.target
housing_df.head()
```

```python
from sklearn.model_selection import train_test_split
X = housing_df.drop('MedHouseVal',axis= 1)
y = housing_df['MedHouseVal']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_squared_error
import time
```

```python
start = time.time()
l_model = LinearRegression()
l_model.fit(X_train,y_train)
end = time.time()

y_pred = l_model.predict(X_test)
r2s = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)

print("r2score ",r2s," MSE ",mse," time ",end-start)
```

```python
start = time.time()
g_model = GradientBoostingRegressor(n_estimators=100)
g_model.fit(X_train,y_train)
end = time.time()

y_pred = g_model.predict(X_test)
r2s = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)

print("r2score ",r2s," MSE ",mse," time ",end-start)
```

```python
start = time.time()
x_model = XGBRegressor()
x_model.fit(X_train,y_train)
end = time.time()

y_pred = x_model.predict(X_test)
r2s = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)

print("r2score ",r2s," MSE ",mse," time ",end-start)
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples=10000,n_features=20,n_redundant=2, n_repeated=4, n_classes=2,random_state=10)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
start = time.time()
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
end = time.time()

y_pred = log_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
cr = classification_report(y_test,y_pred)

print("Accuracy :", accuracy," \nclassification report:\n",cr," \nTime: ",end-start)
```

```python
start = time.time()
gc_model = GradientBoostingClassifier()
gc_model.fit(X_train,y_train)
end = time.time()

y_pred = gc_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
cr = classification_report(y_test,y_pred)

print("Accuracy :", accuracy," \nclassification report:\n",cr," \nTime: ",end-start)
```

```python
start = time.time()
xc_model = XGBClassifier()
xc_model.fit(X_train,y_train)
end = time.time()

y_pred = xc_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
cr = classification_report(y_test,y_pred)

print("Accuracy :", accuracy," \nclassification report:\n",cr," \nTime: ",end-start)
```
